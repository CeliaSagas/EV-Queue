```python
import pandas as pd
pd.set_option('display.max_columns', 500)
import datetime as dt
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
#replace white space in column names with _, strip ending white space

data.columns = data.columns.str.replace(' ', '_')
data.columns = data.columns.str.rstrip()
```

```python
data.dtypes
```

```python
#convert date time columns to date time format 

date_cols = ["Start_Date", "End_Date", "Transaction_Date_(Pacific_Time)"]
duration_cols = ["Total_Duration_(hh:mm:ss)", "Charging_Time_(hh:mm:ss)"]

for col in date_cols:
     data[date_cols] = data[date_cols].apply(pd.to_datetime, errors = "coerce")

for col in duration_cols:
    data[duration_cols] = data[duration_cols].apply(pd.to_timedelta, errors="raise")
   
    
```

```python
#create dataframe of 2019 data 

data_2019 = data[data["Start_Date"] >= "1/1/2019" AND data["Start_Date"] < "1/1/2020"].reset_index(drop=True)
```

```python
#create week column 

data_2020["WEEK"] = data_2020["End_Date"].dt.isocalendar().week
```

```python
data_2020.head()
```

```python
week_2020 = data_2020.groupby(["WEEK","Address_1", "Station_Name"]).count().reset_index()
```

```python
week_2020 = week_2020[["WEEK", "Address_1", "Station_Name", "Fee"]]
```
